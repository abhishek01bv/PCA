```python
plt.figure(figsize=(30,20))
plotnumber=1

for column in df.drop('Bankrupt?', axis=1):
  plt.subplot(10,10,plotnumber)
  sns.boxplot(df[column])
  plotnumber+=1

plt.show()
```

```python
replace_outliers = df.copy()
for column in df.columns:
  q1 = df[column].quantile(0.25)
  q3 = df[column].quantile(0.75)
  iqr = q3 - q1
  upper  = 1+1.5*iqr
  lower = q1-1.5*iqr
  outlier = (df[column]<lower) | (df[column]>upper)
  upper =0
  lower =0
  replace_outliers.loc[outlier, column ] = df[column].median
print(replace_outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2 = scaler.fit_transform(df.drop('Bankrupt?',axis=1))
```

```python
df2 = pd.DataFrame(df2, columns =  df.drop('Bankrupt?',axis=1).columns)
df2
```

```python
df2.isnull().sum()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pc = pca.fit_transform(df2)
```

```python
pca.explained_variance_ratio_
```

```python
for index,value in enumerate(pca.explained_variance_ratio_):
    print("Variance for PCA",index+1,"is",value)
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_),marker="o")
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

```python
pca = PCA(n_components=5) #Object Creation
new_data = pca.fit_transform(df2)
# This will be the new data fed to the algorithm.
principal_Df = pd.DataFrame(data = new_data)
```
